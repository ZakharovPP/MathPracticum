```python
from sklearn.datasets import fetch_openml

images, labels = fetch_openml('mnist_784', return_X_y=True, as_frame=False)
```

```python
def k_means(X, k, max_iter=100, threshold=1e-9):
    N, M = X.shape
    idx = np.random.choice(N, size=k, replace=False)
    centroids = X[idx].copy()
    wcss_old = np.inf
    for _ in range(max_iter):
        X_sq = np.sum(X ** 2, axis=1, keepdims=True)
        C_sq = np.sum(centroids ** 2, axis=1, keepdims=True).T
        XC = -2 * X @ centroids.T
        distances_sq = X_sq + C_sq + XC
        distances_sq = np.maximum(distances_sq, 0)
        distances = np.sqrt(distances_sq)
        labels = np.argmin(distances, axis=1)
        wcss_new = np.sum(np.min(distances_sq, axis=1)) 
        if abs(wcss_old - wcss_new) < threshold:
            break
        wcss_old = wcss_new
        for i in range(k):
            if np.any(labels == i):
                centroids[i] = X[labels == i].mean(axis=0)
    return labels, wcss_new
```

```python
labels_new, wcss = k_means(images, 10, 100, 1e-4)
centroids = np.zeros((10, images.shape[1]))  
for i in range(10):
    centroids[i] = images[labels_new == i].mean(axis=0)
for i in range(10):
    plt.imshow(centroids[i].reshape(28, 28), cmap='Greys_r')
    plt.title(f'Cluster {i} center')
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix
labels = labels.astype(int)
cluster_to_digit = {}
cluster_accuracy = {}
for cluster_id in range(10):
    mask = (labels_new == cluster_id)
    true_in_cluster = labels[mask]
    if len(true_in_cluster) == 0:
        digit = -1
        correct = 0
        total = 0
        accuracy = 0.0
    else:
        counts = np.bincount(true_in_cluster, minlength=10)
        digit = np.argmax(counts)
        correct = counts[digit]
        total = len(true_in_cluster)
        accuracy = correct / total * 100
    cluster_to_digit[cluster_id] = digit
    cluster_accuracy[cluster_id] = (correct, total, accuracy)
    print(f" {cluster_id}   {digit} | "
          f": {correct}/{total} ({accuracy:.1f}%)")
predicted_digits = np.array([cluster_to_digit[cl] for cl in labels_new])
cm = confusion_matrix(labels, predicted_digits, labels=range(10))
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=range(10), yticklabels=range(10))
plt.title('  / ')
plt.xlabel(' ')
plt.ylabel(' ')
plt.show()
total_correct = np.sum(predicted_digits == labels)
total_samples = len(labels)
overall_accuracy = total_correct / total_samples * 100
print(f"\n : {total_correct}/{total_samples} ({overall_accuracy:.1f}%)")
```
